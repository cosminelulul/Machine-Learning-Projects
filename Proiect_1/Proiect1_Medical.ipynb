V. Vizualizarea datelor 

```python
import matplotlib.pyplot as plt  
```

```python
plt.figure(figsize=(6,6))  # Setează dimensiunea figurii
Gender = df['Gender'].value_counts()

plt.pie(
    Gender, 
    labels=['Female', 'Male'],  # Etichete pentru segmente
    autopct='%1.1f%%',         # Afișează procentajul fiecărui segment
    startangle=90              # Începe graficul de la 90 de grade
)

plt.title('Genul')  # Titlul graficului
plt.show()                                # Afișează graficul
```

```python
# Obține numărul de din DataFrame-ul filtrat
Neighbourhood = df['Neighbourhood'].value_counts()
Neighbourhood = Neighbourhood.head(10)

# Selectează 
Neighbourhood = Neighbourhood.sort_values(ascending=True)


plt.barh(Neighbourhood.index, Neighbourhood)  
plt.ylabel('Nr de pacienti din urmatoarele cartiere')
plt.show()

```

```python
Handicap = df['Handcap'].value_counts()
```

```python
plt.bar(Handicap.index,Handicap.values)
plt.ylabel('Nr de pacienti din urmatoarele cartiere')
plt.show()

```
