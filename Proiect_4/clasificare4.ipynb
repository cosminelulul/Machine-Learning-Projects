<h1 style="text-align: justify; font-family: 'TimesNweRoman'; font-size:30px; color:LightBlue"> Împărțirea datelor în set de antrenament și set de test </h1>


```python
X = StudentPlacement_modificat.drop('Placement_Yes',axis=1) 
y = StudentPlacement_modificat['Placement_Yes']
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=21)
```

```python
# --------------------------
# Scalarea dataset-ului folosind MinMaxScaler
# --------------------------

from sklearn.preprocessing import MinMaxScaler

# Crearea obiectului scaler
scaler = MinMaxScaler()

# Învățarea min și max pe baza întregului set X
scaler.fit(X)

# Aplicare scaler pe setul de antrenament
X_train = scaler.fit_transform(X_train)

# Aplicare scaler pe setul de test
X_test = scaler.transform(X_test)

# Transformarea rezultatul în DataFrame pentru lizibilitate, păstrând numele coloanelor
X_train = pd.DataFrame(X_train, columns=X.columns)
X_test = pd.DataFrame(X_test, columns=X.columns)

# Afișare primelor 5 rânduri din seturile scalate
print('========== Antrenare ==========\n')
X_train.head()

```

```python
print('========== Testare ==========\n')

X_test.head()
```

<h2 style="text-align: justify; font-family: 'TimesNewRoman'; font-size:25px; color:LightBlue"> Regresia Logistică </h2>

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, f1_score
from sklearn.linear_model import LogisticRegression

# Setarea dimensiunii figurii și mărimea fontului
plt.rcParams['figure.figsize'] = (8,6)
plt.rcParams['font.size'] = 20

# --------------------------
# Regresie logistică
# --------------------------

logreg = LogisticRegression()

# Antrenarea modelului pe datele de training
logreg.fit(X_train, y_train)

# Prezicerea valorilor pentru setul de test
y_pred_log = logreg.predict(X_test)

# Calcularea acurateții pe setul de training
log_train = round(logreg.score(X_train, y_train) * 100, 2)

# Calcularea acurateții pe setul de test
log_accuracy = round(accuracy_score(y_pred_log, y_test) * 100, 2)

# Calcularea scorului F1
log_f1 = round(f1_score(y_pred_log, y_test) * 100, 2)

# Afișarea rezultatelor
print("Acuratețea pe setul de training :", log_train, "%")
print("Acuratețea modelului          :", log_accuracy, "%")
print("Scorul F1                     :", log_f1, "%")
print("\033[1m--------------------------------------------------------\033[0m")

# Raport de clasificare (precizie, recall, F1 pentru fiecare clasă)
print("Raport de clasificare:\n", classification_report(y_test, y_pred_log))
```

* **Observație!** Modelul de regresie logistică prezintă o acuratețe bună, obținând 90.21 % pe setul de antrenare și 89.73  % pe setul de testare, ceea ce indică o generalizare modesta.  
Scorul F1 de 63.07 % arată un echilibru modest între precizie (0.76–0.91) și recall (0.54–0.79) pentru ambele clase.
indicând *overfitting* și o diferență de performanță între clase.

```python
from sklearn.metrics import ConfusionMatrixDisplay

# -----------------------------
# Matricea de confuzie pentru modelul de regresie logistică
# -----------------------------

plt.rcParams['figure.figsize'] = (10, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 


# Calculare matrice de confuzie
cm = confusion_matrix(y_test, y_pred_log)

# Transformarea matricei într-un DataFrame pentru vizualizare mai clară
cm_df = pd.DataFrame(cm)

# Vizualizare cu Seaborn

sns.heatmap(cm_df, annot=True, fmt='d', cmap='Purples', cbar=False)
plt.title("Matricea de Confuzie - Regresia Logistică")
plt.ylabel("Valori reale")
plt.xlabel("Valori prezise")
plt.show()
```

* **Observație!** Matricea de confuzie arată că modelul de regresie logistică are o performanță mediocra, corect clasificând majoritatea observațiilor. Din totalul de 3000 de cazuri, 2692 au fost identificate corect, rezultând o acuratețe de aproximativ 86%.

```python
from sklearn.metrics import RocCurveDisplay

# -----------------------------
# Curba ROC pentru modelul de regresie logistică
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei ROC pe baza estimatorului (modelului)
RocCurveDisplay.from_estimator(
    logreg,   # modelul antrenat
    X_test,   # datele de test
    y_test,   # valorile reale
)

# Titlul graficului 
plt.title("Curba ROC - Regresie Logistică")

# Afișare figură
plt.show()
```

* **Observație!** Modelul de regresie logistică are o performanță foarte bună în clasificarea datelor, având o valoare AUC de 0.93. Curba ROC este situată aproape de colțul stânga-sus al graficului, ceea ce arată o capacitate ridicată de diferențiere între clasele pozitive și negative.

```python
from sklearn.metrics import PrecisionRecallDisplay

# -----------------------------
# Curba Precizie-Recall pentru modelul de regresie logistică
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei PR pe baza estimatorului (modelului)
PrecisionRecallDisplay.from_estimator(
    logreg,    # modelul antrenat
    X_test,    # datele de test
    y_test,    # valorile reale
)

# Titlul graficului 
plt.title("Curba Precizie-Recall - Regresie Logistică")

# Afișare figură
plt.show()
```

* **Observație!** Modelul de regresie logistică prezintă o performanță modesta conform curbei Precizie–Recall, având o valoare a preciziei medii (AP) de 0.77. Acest rezultat indică faptul că modelul reușește să identifice corect mare parte exemplelor pozitive, dar  în același timp mentina rată mai mare a predicțiilor false pozitive.

```python
from sklearn.model_selection import cross_val_score

# -----------------------------
# Scorul de validare încrucișată
# -----------------------------

# cv=10 -> împărțirea datelot în 10 fold-uri
val_score = cross_val_score(estimator=logreg, X=X_train, y=y_train, cv=10)

# Afișarea acurateții medie a modelului
print("Scorul mediu al modelului: {:.2f} %".format(val_score.mean()*100))

# Afișarea deviației standard a scorurilor pentru a vedea variația între fold-uri
print("Deviația standard: {:.2f} %".format(val_score.std()*100))
```

* **Observație!** În urma validării încrucișate (cross-validation), modelul de regresie logistică a obținut un scor mediu de 90.29  % și o deviație standard de 0.96  %, ceea ce indică o performanță bună și stabilă. Valoarea redusă a deviației standard arată că modelul oferă rezultate consistente pe diferite subseturi ale datelor, fără variații semnificative în acuratețe.

<h2 style="text-align: justify; font-family: 'TimesNewRoman'; color:LightBlue"> 2) Arbore de decizie </h2>

```python
# -----------------------------
# Decision Tree Classifier
# -----------------------------

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, f1_score, classification_report

# Crearea modelul Decision Tree
decision = DecisionTreeClassifier()

# Antrenarea modelului pe datele de training
decision.fit(X_train, y_train)

# Prezicerea valorilor pentru setul de test
y_pred_Decision = decision.predict(X_test)

# Calcularea acurateții pe setul de training
decision_train = round(decision.score(X_train, y_train) * 100, 2)

# Calcularea acuratețeii pe setul de test
decision_accuracy = round(accuracy_score(y_pred_Decision, y_test) * 100, 2)

# Calcularea scorului F1 pe setul de test
decision_f1 = round(f1_score(y_pred_Decision, y_test) * 100, 2)

# Afișarea rezultatelor
print("Acuratețea pe setul de training :", decision_train, "%")
print("Acuratețea modelului           :", decision_accuracy, "%")
print("Scorul F1                      :", decision_f1, "%")
print("\033[1m--------------------------------------------------------\033[0m")

# Raport de clasificare (precizie, recall, F1 pentru fiecare clasă)
print("Raport de clasificare:\n", classification_report(y_test, y_pred_Decision))
```

* **Observație!** Arborele de decizie a învățat perfect datele de training (100% acuratețe), iar pe setul de test performează perfect cu 100% acuratete, indicând ca nu exista *overfitting*.


```python
# -----------------------------
# Matricea de confuzie - Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (10, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Calculare matrice de confuzie
cm = confusion_matrix(y_test, y_pred_Decision)

# Transformarea matriceei într-un DataFrame pentru vizualizare mai clară
cm_df = pd.DataFrame(cm)

# Vizualizare cu Seaborn

sns.heatmap(cm_df, annot=True, fmt='d', cmap='Blues', cbar=False)
plt.title("Matricea de Confuzie - Decision Tree")
plt.ylabel("Valori reale")
plt.xlabel("Valori prezise")
plt.show()
```

```python
# -----------------------------
# Curba ROC - Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei ROC pe baza modelului Decision Tree
RocCurveDisplay.from_estimator(
    decision,       # modelul antrenat
    X_test,    # datele de test
    y_test     # valorile reale
)

plt.title("Curba ROC - Decision Tree")

# Afișare figură
plt.show()
```

* **Observație!** Curba ROC arată că modelul are o capacitate perfecta de discriminare între clase, cu un **AUC de 1.00**, indicând că există o probabilitate de 100% ca modelul să clasifice corect un exemplu pozitiv.


```python
# -----------------------------
# Curba Precizie-Recall -  Decision Tree Classifier
# -----------------------------

plt.rcParams['figure.figsize'] = (14, 8)   # dimensiunea figurii
plt.rcParams['font.size'] = 15 

# Crearea afișajului curbei Precizie-Recall
PrecisionRecallDisplay.from_estimator(
    decision,       # modelul antrenat
    X_test,    # datele de test
    y_test     # valorile reale
)

plt.title("Curba Precizie-Recall - Decision Tree")

#Afișare figură
plt.show()
```

* **Observație1** Curba PR (Precision-Recall) arată că modelul menține un echilibru perfect între **precizie și recall**, cu un **Average Precision (AP) de 1.00**, indicând performanță excellenta în identificarea corectă a exemplelor pozitive.


```python
# -------------------------------------------------------------
# Validare încrucișată (Cross-Validation) pentru Decision Tree
# -------------------------------------------------------------

# cv=10 -> împărțirea datelor în 10 fold-uri
val_score_dtc = cross_val_score(estimator=decision, X=X_train, y=y_train, cv=10)

# Afișarea scorului mediu al modelului pe toate fold-urile
print("Scorul mediu al modelului Decision Tree: {:.2f} %".format(val_score_dtc.mean()*100))

# Afișarea deviației standard pentru a evalua variația scorurilor între fold-uri
print("Deviația standard: {:.2f} %".format(val_score_dtc.std()*100))
```

* **Observație!** Validarea încrucișată a modelului Decision Tree a evidențiat o acuratețe medie de **99.97 %**, cu o deviație standard de **0.06%**, sugerând o performanță excelenta și o variabilitate relativ scăzută între diferitele fold-uri, indicând o generalizare superiora pe seturile de date.
